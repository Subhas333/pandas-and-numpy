```python
# Filter rows — like NumPy boolean filtering
survivors = df[df["Survived"] == 1]
print("\nTotal survivors:", len(survivors))
```

```python
# Filter with multiple conditions
young_survivors = df[(df["Survived"] == 1) & (df["Age"] < 18)]
print("Young survivors (under 18):", len(young_survivors))
```

```python
# Survival rate by gender
print("=== Survival rate by gender ===")
print(df.groupby("Sex")["Survived"].mean().round(2))

```

```python
print("\n=== Survival rate by passenger class ===")
print(df.groupby("Pclass")["Survived"].mean().round(2))
```

```python
# Average age by class
print("\n=== Average age by class ===")
print(df.groupby("Pclass")["Age"].mean().round(1))
```

```python
# Strategy 1 — fill with median (best for Age)
medium_age = df["Age"].median()
df["Age"] = df["Age"].fillna(medium_age)
```

```python
df.head()
```

```python
#for cabin
print(df['Cabin'].unique())
df=df.drop('Cabin',axis=1)
```

```python
df
```

```python
#  fill with most common value (for Embarked)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.head()
```

```python
# Verify — no more missing values in these columns
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
# Create a new column — family size
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1   # +1 for themselves

# Create age category
def age_group(age):
    if age < 18:
        return "Child"
    elif age < 60:
        return "Adult"
    else:
        return "Senior"

df["AgeGroup"] = df["Age"].apply(age_group)

# Check results
print(df[["Name", "Age", "AgeGroup", "FamilySize"]].head(10))
```

```python
# Average ticket of fair
avg_ticket = round(df['Fare'].mean(),2)
print(f"Average ticket price: ${avg_ticket}")
```
