```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/master/World%20Happiness%20Data/2020.csv"

df = pd.read_csv(url)

print("Shape:", df.shape)
df.head()
```

```python
# See all column names
print("Columns:\n", df.columns.tolist())

# Data types
print("\nData types:\n", df.dtypes)

# Missing values
print("\nMissing values:\n", df.isnull().sum())

# Basic statistics
print("\nBasic stats:")
df.describe().round(2)
```

```python
# Top 10 happiest countries
print("=== Top 10 Happiest Countries ===")
top10 = df.nlargest(10, "Ladder score")[["Country name", "Ladder score", "Logged GDP per capita"]]
print(top10.to_string(index=False))

# Bottom 10
print("\n=== 10 Least Happy Countries ===")
bottom10 = df.nsmallest(10, "Ladder score")[["Country name", "Ladder score", "Logged GDP per capita"]]
print(bottom10.to_string(index=False))

# Visualise top 10
plt.figure(figsize=(10, 5))
sns.barplot(data=top10, x="Ladder score", y="Country name", palette="Greens_r")
plt.title("Top 10 Happiest Countries (2020)")
plt.xlabel("Happiness Score (Ladder score)")
plt.tight_layout()
plt.show()
```

```python
# Scatterplot — GDP vs Happiness
plt.figure(figsize=(9, 6))
sns.scatterplot(data=df,
                x="Logged GDP per capita",
                y="Ladder score",
                hue="Ladder score",
                palette="YlGn",
                size="Ladder score",
                sizes=(40, 200),
                legend=False)

# Label top 5 happiest countries
for _, row in df.nlargest(5, "Ladder score").iterrows():
    plt.annotate(row["Country name"],
                 (row["Logged GDP per capita"], row["Ladder score"]),
                 fontsize=8, xytext=(5, 3), textcoords="offset points")

# Label bottom 5
for _, row in df.nsmallest(5, "Ladder score").iterrows():
    plt.annotate(row["Country name"],
                 (row["Logged GDP per capita"], row["Ladder score"]),
                 fontsize=8, xytext=(5, -10), textcoords="offset points")

plt.title("GDP per Capita vs Happiness Score (2020)")
plt.xlabel("Logged GDP per Capita")
plt.ylabel("Happiness Score")
plt.show()

# Correlation value
corr = df["Logged GDP per capita"].corr(df["Ladder score"])
print(f"Correlation between GDP and Happiness: {corr:.2f}")
```

```python
# Correlation of all numeric factors with Ladder score
numeric_df = df.select_dtypes(include=np.number)
correlations = numeric_df.corr()["Ladder score"].drop("Ladder score").sort_values(ascending=False)

print("Correlation with Happiness Score (Ladder score):")
print(correlations.round(2))

# Horizontal bar chart
plt.figure(figsize=(10, 6))
colors = ["#1D9E75" if v > 0 else "#E24B4A" for v in correlations]
correlations.plot(kind="barh", color=colors)
plt.title("What Factors Correlate Most with Happiness? (2020)")
plt.xlabel("Correlation with Ladder Score")
plt.axvline(0, color="black", linewidth=0.8)
plt.tight_layout()
plt.show()
```

```python
# Distribution with KDE curve
plt.figure(figsize=(9, 5))
sns.histplot(df["Ladder score"], bins=25, kde=True,
             color="#534AB7", edgecolor="white")

mean_val = df["Ladder score"].mean()
median_val = df["Ladder score"].median()

plt.axvline(mean_val, color="red", linestyle="--",
            label=f"Mean: {mean_val:.2f}")
plt.axvline(median_val, color="orange", linestyle="--",
            label=f"Median: {median_val:.2f}")

plt.title("Distribution of World Happiness Scores (2020)")
plt.xlabel("Happiness Score (Ladder score)")
plt.ylabel("Number of Countries")
plt.legend()
plt.show()

print(f"Skewness: {df['Ladder score'].skew():.2f}")
print(f"Mean:     {mean_val:.2f}")
print(f"Median:   {median_val:.2f}")
print(f"Std Dev:  {df['Ladder score'].std():.2f}")
```

```python
# Select key columns only
cols = ["Ladder score",
        "Logged GDP per capita",
        "Social support",
        "Healthy life expectancy",
        "Freedom to make life choices",
        "Generosity",
        "Perceptions of corruption"]

# Pairplot — all relationships at once
sns.pairplot(df[cols], diag_kind="kde",
             plot_kws={"alpha": 0.5, "color": "#534AB7"},
             diag_kws={"color": "#534AB7"})
plt.suptitle("Pairplot — World Happiness Factors 2020", y=1.02)
plt.show()

# Correlation heatmap
plt.figure(figsize=(9, 7))
sns.heatmap(df[cols].corr().round(2),
            annot=True, fmt=".2f",
            cmap="coolwarm", center=0,
            square=True, linewidths=0.5)
plt.title("Correlation Heatmap — World Happiness 2020")
plt.tight_layout()
plt.show()
```
